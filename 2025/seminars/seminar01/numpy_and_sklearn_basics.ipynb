## Introduction to matplotlib

```python
# lets make some plots!
```

```python
data = np.random.normal(loc=25, scale=2, size=(1000))
thetas = []
for i in range(1, data.shape[0]):
    thetas.append(data[:i].mean())
```

```python
plt.figure(figsize=(10, 5))
plt.title(r'$\Theta$ estimation')
plt.grid(linestyle='-', linewidth=0.5)
plt.plot(thetas, label='Mean estimation')
plt.axhline(25, color='red', label=r'$\Theta$')
plt.legend()
plt.show()
```

```python
def f(x):
    return x ** 2

def g(x):
    return x ** 3

x = np.linspace(-1, 1)

sub1 = plt.subplot(1, 2, 1)
sub1.set_title('f(x)')
sub1.plot(x, f(x))

sub2 = plt.subplot(1, 2, 2)
sub2.set_title('g(x)')
sub2.plot(x, g(x))
plt.show()
```

```python
dist1 = np.random.normal(loc=(-0.5, -0.5), size=(100, 2))
dist2 = np.random.normal(loc=(0.5, 0.5), size=(100, 2))
```

```python
plt.title('Two distributions')
plt.scatter(dist1[:, 0], dist1[:, 1])
plt.scatter(dist2[:, 0], dist2[:, 1])
plt.show()
```

## Introduction to sklearn

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_digits
```

```python
X, y = load_digits(return_X_y=True)
```

```python
X.shape
```

```python
rows = 2
columns = 4

plt.gray()
for row in range(rows):
    for column in range(columns):
        index = columns * row + column + 1
        sub = plt.subplot(rows, columns, index)
        sub.matshow(X[index].reshape(8, 8))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=218)
```

```python
neigh.fit(X_train, y_train)
```

```python
accuracy_score(y_test, neigh.predict(X_test))
```

## Iris dataset

```python
from sklearn.datasets import load_iris
```

```python
# more info https://en.wikipedia.org/wiki/Iris_flower_data_set
dataset = load_iris()
print(dataset.DESCR)
```

```python
X, y = dataset.data, dataset.target
```

```python
# projection
from mpl_toolkits.mplot3d import Axes3D
plt.winter()

fig = plt.figure(figsize=(8, 8))
ax = Axes3D(fig, auto_add_to_figure=False)
fig.add_axes(ax)

ax.scatter(X[:, 0], X[:, 1], X[:, 3], c=y, marker='x')
ax.set_xlabel(dataset.feature_names[0])
ax.set_ylabel(dataset.feature_names[1])
ax.set_zlabel(dataset.feature_names[2])
plt.show()
```

```python
import pandas as pd
df = pd.DataFrame(
    np.concatenate((X, y[:, None]), axis=1),
    columns=dataset.feature_names + ['target label'],
)
df.head()
```

```python
import plotly.express as px

fig = plt.figure(figsize=(8, 8))
fig = px.scatter_3d(df, x='sepal length (cm)', y='sepal width (cm)', z='petal width (cm)',
                    color='target label')
fig.show()
```
